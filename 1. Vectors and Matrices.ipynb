## 3. Vector Multiplication (The Dot Product)
###  calculate the sum of the multiplied elements of two vectors of the same length to give a  scalar. 
### If "a"  and "b" are two vectors thwen then their dot product is a scalar "c"
### c = a.b. = (a1 * b1 + a2 * b2 + a3 * b3)  where a =(a1,a2,a3) and b =(b1,b2,b3)
## The Dot product c = a.b can also be represented as $c_{i,j}$  =  $\sum\limits_{i}^{n}$ $a_{i}$$b{i}$
### <u> One application of dot product  it is  used to calculate weighted sum of inputs in a neuron </u>
#### <u> Another dot product application is in matrix factorization or decompositon techniques to obtain its decomposed elements</u>

```python
a = array([50,90,13])
b  = array([1,8,6])
print("The one dimensional of column vector a  ", a.shape)  # 1 dimensional column vector "a" 
                                                            # but a as (3,) cant be a as dimension of  3*1 
c = a.dot(b)  ## c = [1*2 + 2*3 +3*4]    ### dot product with 1 dimensional column  vectors 
print("The  dot product of a and b is c= ",c)
```

### 4. Vector Scalar Multiplication

### Also the transpose of transpose of a matrix A is the maatrix A itself
### $ (A^{T})^T = A  $
