### If A is m * n and B is n * p then C = A.B = m * p

###  <u>****  In the dot product of matrix multiplication, calculate the dot product between each row in matrix A with each column in matrix B. </u>

```python
import numpy as np 
A = np.array([[0,1,2],[1,2,3],[2,3,4]])
B= np. array([[1,-2],[-1,0],[2,-1]])
print("The matrix A is")
print(A) 
print("The matrix B is ")
print(B)
C = A.dot(B)                     ####   ([1*2+2*4, 1*3+2*2, 1*4 +2*1]
print("The dot product of matrix A of order 3*3 and B of order 3*1 is C " )
print( C ," = C is matrix of order 3*1")   
print("The dimension of new matrix C is ", C.shape)
```

## Orthogonal Matrix - a square matrix whose rows and columns are orthonormal unit vectors i.e. are perpendicluar and have a length of 1. Also ,if two vectors are orthonormal then they are orthogonal .
## $\begin {bmatrix} 1 & 0 \\ 0 & -1\end{bmatrix}$

### Two vectors are orthogonal when their dot product equals zero. The length of each vector is 1 then the vectors are called orthonormal because they are both orthogonal and normalized.
## w.x = 0   or w.$x^T = 0 $

## If the transpose of a square matrix with real entries is equal to its inverse, then the matrix is called an orthogonal matrix.

## Orthogonal Matrix Properties --> the following properties are important while processing data in the context of machine and deep learning
### 1. Denoted by Q 
### 2. Multiplication by orthogonal matrix preserves length
### 3. A matrix is orthogonal if the dot product of matrix  and its transpose is equal to identity matrix  
### $Q.Q^T = Q^T.Q = I $
### 4. A matrix is orthogonal if its transpose is equal to its inverse
### $ Q^T = Q^{-1}$

### 5. The determinant of an orthogonal matrix is either +1 or -1.

### 6. The product of two orthogonal matrices is also an orthogonal matrix.

### 7. The eigenvalues of an orthogonal matrix have absolute value 1.

###  8. The singular values of an orthogonal matrix are all equal to 1.

###  9. The Frobenius norm of an orthogonal matrix is equal to the square root of its order.

### Orthogonal Matrix Property Verification

```python
from numpy import array
from numpy.linalg import inv
Q = array([[-1,0],[0,1]])
print(Q)
print("The transpose of matrix Q is ")
print(Q.T)

I  = Q.dot(Q.T)  ## identity equivalence
print(I)

IQ= inv(Q)   #  inverse of a matrix using inv keyword

print(IQ)
```

## The Rank of a Matrix    
### <u>You can think of an r x c matrix as a set of r row vectors, each having c elements; or you can think of it as a set of c column vectors, each having r elements.</u>

### The rank of a matrix is defined as
#### (a) the maximum number of linearly independent column vectors in the matrix or
#### (b) the maximum number of linearly independent row vectors in the matrix. Both definitions are equivalent.

### For an r x c matrix,
### •	If r is less than c, then the maximum rank of the matrix is r.
### •	If r is greater than c, then the maximum rank of the matrix is c.
### The rank of a matrix would be zero only if the matrix had no elements. If a matrix had even one element, its minimum rank would be one
 https://stattrek.com/matrix-algebra/matrix-rank 

### *** Below example clarifies that the rank of a matrix isn't the number of dimensions of a matrix but a total number of linearly independent directions

### <u> Once the matrix rank is known then it is be posible to eliminate other derived rows or columns from the matrix thus reducing the dimensions of the dataset </u>

### An analogy is --> In a relational database a Derived attribute is an attribute that can be obtained from other attribute for example age is an attribute  derived from date of birth, so having stored date of birth  in a relational table there is no need to store the age attribute. In similar way the linearly  dependent rows or columns makes the dataset reducible as plays an importn role in dimensionality redcution once we know the rank of a matrix. 


```python
from numpy import array
from numpy.linalg import matrix_rank 

A = array([[0,0],[0,0]])
rankA = matrix_rank(A)   # rank 0
print("The matrix A is ")
print(A)
print("The rank of matrix A is ",rankA)

B = array([[1,2],[1,2]])
rankB = matrix_rank(B)   # row 2 is scalar multiple of row 1 , so rank 1
print("The matrix B is")
print(B)
print("The rank of matrix B is ",rankB)


C = array([[1,2],[3,4]])
rankC = matrix_rank(C)    # row1 and row2  are independent of each other so rank 2
print("The matrix C is ")
print(C)
print("The rank of matrix C is ",rankC)

D= array([[1,2],[3,4],[4,6]])  # row 3 is linear combination of row1 and row2  so rank 2
rankD = matrix_rank(D)
print("The matrix D is ")
print(D)
print("The rank of matrix D is ",rankD)

E = array([[2,3,4,-1],[5,2,0,-1],[-4,5,12,-1]])
rankE = matrix_rank(E)
print("The matrix E is ")
print(E)
print("The rank of matrix E is ",rankE)

```
