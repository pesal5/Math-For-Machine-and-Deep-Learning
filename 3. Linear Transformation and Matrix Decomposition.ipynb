###  Linear Transformation Of a  Matrix
###  Suppose there is a matrix $\begin{bmatrix} 2 & 3 \\ 2 & 1\end{bmatrix}$ and a vector$ \begin{bmatrix} x  \\ y\end{bmatrix} $
#### The matrix transforms the vector (x,y ) to a new vector (2x+3y,2x+y) as follows 
$\begin{bmatrix} 2 & 3 \\ 2 & 1\end{bmatrix} \begin{bmatrix} x  \\ y\end{bmatrix}$ = $ \begin{bmatrix} 2x + 3y \\ 2x +y\end{bmatrix}  $
### i.e Linear transformation is a function(T) that maps a vector(V) to some other vector(W) 
### i.e $T : V --> W $ hecne  T : (x,y )  --> (2x+3y,2x+y)
### lets see to what values does (x,y) gets mapped to by a matrix$\begin{bmatrix} 2 & 3 \\ 2 & 1\end{bmatrix}$
 
 
 | Current vector space(X,Y) | new mapped vector space= (2x+3y,2x+y)| 
| --- | --- | 
| (0,1) | (3,1)| 
| (2,2) | (10,6)| 
| (3,2) | (12,8)| 
| (3,5) | (21,11)|


### The Linear transformation of vector (x,y) to a new vector space (2x+3y,2x+y) as shown in above table

```python
import matplotlib.pyplot as plt
v1 = [3,5]
v11 = [21,11]
v2 = [3,2]
v22 =[12,8]
orgn = [0,0]
figr, axes = plt.subplots()
axes.set_xlim(-5,25)
axes.set_ylim(-5,15)
axes.quiver(orgn[0],orgn[0],v1[0],v1[1],angles='xy',scale_units = 'xy', scale =1, color = 'b')
axes.quiver(orgn[0],orgn[0],v11[0],v11[1],angles='xy',scale_units = 'xy', scale =1)
axes.quiver(orgn[0],orgn[0],v2[0],v2[1],angles='xy',scale_units = 'xy', scale =1, color = 'g')
axes.quiver(orgn[0],orgn[0],v22[0],v22[1],angles='xy',scale_units = 'xy', scale =1, color = 'r')
plt.show()
```

### It can be seen that the vector (3,5) blue arrow in the 2d plane has been linearly  transformed to (21,11) as denoted by the black arrow, and the  vector (3,2) green arrow in the 2d plane has also been linearly transformed to (12,8) by the matrix A.

# Eigen Value and Vectors 
### In linear transformation  of the vector by a certain function there are some vectors whose magnitude is changed but not the direction. Such vectors are called eigen vector and the values by which the eigen vector are scaled is called eigen values. 

### In the above example of graph , the  vector (3,2) in green arrow in the  2d plane has been linearly transformed to (12,8) by the matrix $\begin{bmatrix} 2 & 3 \\ 2 & 1\end{bmatrix}$ but the new vector (12,8) direction is exaclty the  same as the original vector. Hence the vector (3,2) is an eigen vector and 4 is its eigen value becaus the new mapped vector has just scaled by 4 to (12,8)

### Hence we can say that the matrix A = $\begin{bmatrix} 2 & 3 \\ 2 & 1\end{bmatrix}$  has one of its eigen values as 4 and its eigen vector as $\begin{bmatrix} 3 \\ 2 \end{bmatrix}$ 

### This concept has diverse applications in the field of data science and machine learning.
### Eigen values and eigen vectors of a matrix are  the basis on which computers are able to differentiate faces, google's page rank algorithm operates , document retrieval , dimensionality reduction can be done and  is an important concept in the filed of machine and deep learningetc etc 

## Matrix Decomposition - A matrix can be decomposed/factorized into its smaller parts from which interesting patterns and properties about the matrix can be found out. Its like a prime factorization of a big number from which important  properties of a number can be drawn out.

### There are various applicaitons of matrix decomposition --> They make calculation of a big matrix easier, dimensionality reduction technique (like PCA)is another application  where matrix factorization comes into play.

### Eigen Decomposition - A square matrix can be decomposed into its contituent eigenvectors and values using eigen decomposition 
### If A is a matrix , v its eigen vector and $\lambda$ the eigen values then 
### A.x = $\lambda$.x 
### Below code decomposes a matrix into its eigen vectors and values

```python
import numpy as np
A = np.array([[2,3],[3,1]])
print("The matrix A is")
print(A)

egn_val, en_vctr = np.linalg.eig([[2,3],[3,1]])
print("The eigen values of the matrix A are",egn_val)
print("The eigen vectors of the matrix A are")
print(en_vctr )
```

#### The 1st eigen value 4.54138127 has a corresponding eigen vector in column  vector form                                      [ 0.76301998, 0.6463749 ]

## Confirm eigen decompostion equation 
### A.x = $\lambda$.x 

```python
import numpy as np
from numpy.linalg import eig
A = np.array([[2,3],[3,1]])
print(A)
egn_val, en_vctr  = eig(A)
Left_Av0 = A.dot(en_vctr [:,0])  ## First eigen vector  vectors[:,0] 
print(Left_Av0)


Right_lamdv0 = en_vctr [:,0].dot(egn_val[0])
print("The values of AV = Left_Av0 is ", Left_Av0 ,"and values of lambda * V = Right_lamdv0 =  ",Right_lamdv0,"are same" )
```

#### Original Matrix Reconstruction using eigen values and vectors
### $A = V.\lambda.V^{-1}$

```python
from numpy import diag
from numpy.linalg import inv
from numpy import array
from numpy.linalg import eig
# define matrix
A = np.array([[2,3],[3,1]])
# factorize the matrix into eigen vectors and values
egn_val, en_vctr  = eig(A)
# create matrix from eigenvectors
V = en_vctr
# create inverse of eigenvectors matrix
V_I = inv(en_vctr)
# create diagonal matrix from eigenvalues
Lambda = diag(egn_val)
# reconstruct the original matrix from eigenvalues and eigenvectors
Reconstrc_Mtrx = V.dot(Lambda).dot(V_I)
print("The original matrix A is")
print(A)
print("The reconstructed matrix is")
print(Reconstrc_Mtrx)
```

### Significance of Eigen decomposition
### The eigen vectors associated with the highest eigen value retains the maximum amout of information contained in the original matrix. Hence this decomposed matrix/eigen vector can be used as  a model to represent the big original matrix thus reducing the complexity of the original matrix in an efficient way. This concept is widely used in Dimensionality Reduction Technique.

## Singular Value Decomposition (SVD)

## SVD  decompositon is valid for rectagular as well as sqaure matrices and  based on the  eigen value and eigen vector concept  but this decomposition is more stable than eigen decomposition

##  A = $U.\sum $.$V^{T}$   where A is a n * m matrix, U m*m matrix, $\sum $ is a m * n diagonal matrix and $V^{T}$ is transpose of n * n matrix 

### <u> the diagonal values of the matrix $\sum $ gives singular values / sqrt( eigen values  of A), column values of U matrix  gives left singular vector of A   and column values of V matrix  gives  right singular vector of A   </u>

## Application Of SVD--> Data compression, data Reduction ,Denoising, Least Square Regression, Image compression, inverse calculation

##  A = $U.\sum $.$V^{T}$   where A is a m * n matrix, U m*m matrix, $\sum $ is a m * n diagonal matrix and $V^{T}$ is transpose of n * n matrix 

### U and V are orthogonal matrices, $\sum $ is a diagonal matrix 

### The composed values from svd() function returns  $\sum$  as a singular vector, V as $V^T$ and U as m*m matrix where $U = eigen \; vectors \;of\; A.A^{T} , V = eigen \;vectors\; of\; A^{T}.A \;and \sum =  square \;\;root \;\; of  \;eigen \;values\; of \; A.A^{T} \;and \; A^{T}.A $
